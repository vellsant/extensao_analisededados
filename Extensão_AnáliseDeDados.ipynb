```python
# Importando biblioteca Pandas no Python
import pandas as pd

```

```python
#Cria o Dataframe com as informações no CSV. 
combustiveis_df = pd.read_excel("/content/ca-2021-02.xlsx")
```

```python
display(combustiveis_df)
```

```python
display(combustiveis_df.head(10))
```

```python
#Número de linhas e colunas
print(combustiveis_df.shape)

#Número de colunas, apenas
print(combustiveis_df.shape[0])
```

```python
#Demonstra as colunas e a tipagem de dados de cada uma.
print(combustiveis_df.info())
```

```python
#Estatísticas básicas
print(combustiveis_df.describe())
```

```python
#Filtrar por coluna
display(combustiveis_df['Revenda'])
```

```python
#Novo dataframe apenas com as colunas que queremos
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#Exibe a quarta linha
display(ca_df.loc[3])
```

```python
#Exibir da décima à vigésima linha
display(ca_df.loc[9:19])
```

```python
#Criar um dataframe Gas_df contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda) somente com combustível GASOLINA.

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)
        
```

```python
#Mostra o maior valor de venda do Dataframe gas_df
display(gas_df['Valor de Venda'].max())
```

```python
#Como mostrar o posto e município com valor de venda mais alto?
display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```
