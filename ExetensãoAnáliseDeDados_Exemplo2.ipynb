```python
#Importando a biblioteca Pandas
import pandas as pd
```

```python
#Criando o Dataframe 'combustiveis_df'
combustiveis_df = pd.read_excel("/content/ca-2021-02.xlsx")
display(combustiveis_df)
```

```python
#Inserção simples de dados
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
```

```python
#Criar uma coluna "OBS" que tenha escrito "MELHOR CIDADE" quando a coluna MUNICIPIO for igual a SANTOS
combustiveis_df['OBS'] = ["MELHOR CIDADE" if municipio == 'SANTOS' else '' for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SANTOS', 'PRAIA GRANDE']),['Municipio', 'OBS']])
```

```python
#Preenchendo uma coluna 'Valor de Venda - Status' quer verifica:
#Valor de Venda > 6,5 = CARO
#Do contrário, BARATO

import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'CARO', 'BARATO')
display(combustiveis_df[['Municipio','Revenda','Valor de Venda', 'Valor de Venda - Status']])
```

```python
#Calcular postos de gasolina por habitante que temos na amostragem de combustiveis de nov/21
num_habitantes_df = pd.read_csv("/content/ibge_num_habitantes_estimado.csv", sep=";")
display(num_habitantes_df)
```

```python
#Alterar o nome da coluna 'Estado' no dataframe 'num_habitantes_df'
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#Fazer um MERGE entre dois Dataframes
colunas=['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df,  how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna totalmente vazia
merge_df.dropna(axis='columns',inplace=True)
print(merge_df.info())
```

```python
#Destruir colunas específicas
colunas=['Regiao - Sigla','Nome da Rua', 'Numero Rua', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status']

merge_df.drop(labels=colunas, axis=1, inplace=True)
```

```python
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
#Deletar as linhas duplicadas
merge_df.drop_duplicates(inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
merge_df.drop(labels='OBS',axis=1,inplace=True)
```

```python
#Agrupar e contar quantos postos tem nas cidades
postos_por_municipio_df=merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.reset_index(inplace=True)
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1,inplace=True)
postos_por_municipio_df.rename(columns={"Revenda":"Numero de Postos"}, inplace=True)

display(postos_por_municipio_df)
```

```python
print(postos_por_municipio_df.info())
```

```python
#Criar uma nova coluna com a média de postos por habitante

postos_por_municipio_df['Habitantes por Posto'] = postos_por_municipio_df['NumHabitantes2021']/postos_por_municipio_df['Numero de Postos']


display(postos_por_municipio_df)
```

```python
#Implementando gráficos
#Importando a biblioteca matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Gerando um Histograma do Valor de Venda 
plt.hist(combustiveis_df['Valor de Venda'])
#Adciona título ao gráfico
plt.title("Distribuição do preço dos combustíveis - Nov/2021")
#Adiciona legenda aos eixos
plt.xlabel("Preço (BRL)")
plt.ylabel("Quantidade de Coletas")

#Traça a linha com o preço médio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle="dashed", linewidth=1)

#Exibe ("plota") o gráfico
plt.show()
```

```python
#Segundo gráfico
media_combustiveis_df = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(media_combustiveis_df)
```

```python
#definir área do gráfico
#plt.figure(figsize=(10,5))

#Plotar gráfico
media_combustiveis_df.plot(kind="barh", color = "red", title="Média de preços por combustível")
plt.xlabel("Preço em BRL/L")
plt.ylabel("Tipo de Combustível")

#Cria grid
plt.grid()

#Remover as linhas do gráfico
sns.despine()

plt.show
```

```python
  #Personalizando a planilha de Excel
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment
```

```python
#media_combustiveis_df.reset_index(inplace=True)
display(media_combustiveis_df)

excel = "/content/porlitro.xlsx"
media_combustiveis_df.to_excel(excel,"Sumário")
```

```python
#abre o excel no openpyxl
wb = load_workbook(excel)

#Pegar a planilha certa, utilizando o Sheet Name
ws = wb['Sumário']

#Pintar o cabeçalho de cinza
cinzinha = PatternFill("solid", fgColor="808080")
ws['A1'].fill = cinzinha
ws['B1'].fill = cinzinha

#Onde o preço do combustível for maior ou igual a 6.5, pinta a fonte de vermelho e deixa em negrito.
MAX_ROW = ws.max_row
num_linha = 2

while(num_linha<=MAX_ROW):
  coord = 'B'+ str(num_linha)
  if ws[coord].value >= 6.5:
    ws[coord].font = Font(bold=True, color="FF0000")
  num_linha += 1
#salvar o excel
wb.save(excel)
```
