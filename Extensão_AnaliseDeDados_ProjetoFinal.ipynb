```python
#Importando as bibliotecas que serão utilizadas na análise.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Criando um Dataframe com a Base de Dados "games.csv"
games_raw_df = pd.read_csv("/content/games.csv")
display(games_raw_df)
```

Nesta seção iremos criar novos Dataframes apenas com os dados que nos interessam para análise, ou seja, excluiremos algumas colunas.

```python
#Visualiza quais são as colunas presentes no Dataframe "games_raw_df" com a função info()
print(games_raw_df.info())
```

```python
#Cria um Dataframe contendo apenas as colunas "rated", "victory_status", "winner", "increment_code", "white_rating", "black_rating", "opening_eco", "opening_name" e "opening_ply"
games_treated_df = games_raw_df.drop(labels=['id','created_at','last_move_at','turns','white_id','black_id','moves'], axis=1,inplace=False)
print(games_treated_df.info())
```

```python
#Cria um Dataframe apenas com as partidas em que as peças brancas venceram
games_white_win_df = games_treated_df.loc[games_treated_df['winner'] == 'white']
display(games_white_win_df)
print(games_white_win_df.describe())
```

```python
#Cria um Dataframe apenas com as partidas em que as peças pretas venceram
games_black_win_df = games_treated_df.loc[games_treated_df['winner'] == 'black']
display(games_black_win_df)
print(games_black_win_df.describe())
```
